```python
df = df[['Date', 'NH3', 'NO2', 'CO', 'PM2.5', 'Temp', 'Pressure', 'Humidity', 'O3']]
```

```python
df['Date'] = df['Date'].apply(lambda x: str(x).split('.')[0])
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d|%H:%M:%S', errors='coerce')
```

```python
df.head()
```

```python
df[df['Date'].isna()]
```

```python
df[df.isna().any(axis=1)]
```

```python
df.loc[78688]
```

```python
df.dropna(how='any', inplace=True)
```

```python
df['NH3'] = df['NH3'].astype(float)
df['NO2'] = df['NO2'].astype(float)
df['CO'] = df['CO'].astype(float)
df['PM2.5'] = df['PM2.5'].astype(float)
df['Temp'] = df['Temp'].astype(float)
df['Pressure'] = df['Pressure'].astype(float)
df['Humidity'] = df['Humidity'].astype(float)
df['O3'] = df['O3'].astype(float)
```

```python
df[df.isna().any(axis=1)]
```

```python
df.dropna(how='any', inplace=True)
```

```python
df.plot(x='Date', subplots=True, figsize=(15, 10));
```

```python
from datetime import datetime as dt
```

```python
start_period = dt.strptime("2020-11-21 00:00:00", '%Y-%m-%d %H:%M:%S')
end_period = dt.strptime("2021-07-01 00:00:00", '%Y-%m-%d %H:%M:%S')
df = df[(df['Date'] > start_period) &  (df['Date'] < end_period)]
```

```python
df.plot(x='Date', subplots=True, figsize=(15, 10));
```

```python
start_period = dt.strptime("2020-11-21 00:00:00", '%Y-%m-%d %H:%M:%S')
end_period = dt.strptime("2021-04-01 00:00:00", '%Y-%m-%d %H:%M:%S')
df = df[(df['Date'] > start_period) &  (df['Date'] < end_period)]
```

```python
df.plot(x='Date', subplots=True, figsize=(15, 10));
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].plot(df['Date'], df['Temp'], c='red')
ax[0, 0].set_title('Temperature, C')

ax[0, 1].plot(df['Date'], df['Pressure'], c='blue')
ax[0, 1].set_title('Pressure, mPa')

ax[1, 0].plot(df['Date'], df['NO2'], c='green')
ax[1, 0].set_title('NO2, g/m3')

ax[1, 1].plot(df['Date'], df['CO'], c='grey')
ax[1, 1].set_title('CO, g/m3')

for axis in ax.ravel():
    axis.tick_params('x', labelrotation=45)

fig.suptitle("Monitored data")
plt.tight_layout()
plt.show()
```
